```python
Esolv_ref = 5.22901
Errores = numpy.zeros((len(Energias)-1))
for i in range(len(Energias)-1):
    Errores[i] = (abs(Energias[i] - Energias[i+1])/Esolv_ref)*100
```

```python
graph_1 = plt.figure(figsize = (12,6))
plt.style.use('seaborn-white')
plt.plot(array_2, Energias, linewidth = 2)
plt.title('Potencial 'r'$\phi{(\mathbf{x})}$' '\n', fontsize = 22)
plt.xlabel(u'\n N√∫mero de im√°genes', fontsize = 18)
plt.ylabel('Valor calculado del potencial \n', fontsize = 18)
plt.xlim(20,62)
plt.xticks(numpy.arange(21,63,2))
plt.axhline(Esolv_ref, color = 'r', linestyle = '--', linewidth = 2)
plt.grid(linestyle = '-.', color = 'k')
plt.show()
graph_1.savefig('Potencial.pdf', bbox_inches = 'tight')
```

```python
graph_2 = plt.figure(figsize = (12,6))
plt.style.use('seaborn-white')
# plt.semilogy(array_1, Errores)
plt.semilogy(array_1, Errores, linewidth = 2)
plt.title(u'Diferencia relativa entre valores de N \n (escala logar√≠tmica)\n', fontsize = 22)
plt.xlabel(u'\n N√∫mero de im√°genes', fontsize = 18)
plt.ylabel('Diferencia porcentual entre \n N consecutivos \n', fontsize = 18)
plt.axhline(Errores[10], color = 'red', linestyle = '--', linewidth = 2)
plt.xticks(numpy.arange(21,61,2), fontsize = 12)
plt.yticks(dummy_y_ticks, log_y_ticks, fontsize = 12)
plt.grid(linestyle = '-.', color = 'k')
plt.show()
graph_2.savefig('Error_log.pdf', bbox_inches = 'tight')
```
